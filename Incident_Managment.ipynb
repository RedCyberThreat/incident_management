**Basic Log Analysis**

```python
import ollama

log_data = """
Failed login attempt from IP 192.168.1.10
Multiple retries detected
"""

response = ollama.chat(
    model='llama3',
    messages=[
        {"role": "system", "content": "You are a cybersecurity analyst."},
        {"role": "user", "content": f"Analyze this log and explain if it's suspicious:\n{log_data}"}
    ]
)

print(response['message']['content'])
```

**Incident Severity Classification**

```python
import ollama

incident = """
User downloaded unknown executable and system behavior changed.
"""

prompt = f"""
Classify the severity (Low, Medium, High) and explain why:

Incident:
{incident}
"""

response = ollama.chat(
    model='llama3',
    messages=[
        {"role": "user", "content": prompt}
    ]
)

print(response['message']['content'])
```

**Automated Incident Report Generator**

```python
import ollama

incident_data = """
Phishing email detected targeting finance department.
User clicked link but no credentials entered.
"""

prompt = f"""
Generate an incident report with:
- Summary
- Impact
- Actions Taken
- Recommendations

Incident:
{incident_data}
"""

response = ollama.chat(
    model='llama3',
    messages=[
        {"role": "user", "content": prompt}
    ]
)

print(response['message']['content'])
```
